```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

temp = [20,25,30,35,40]
sale = [13,21,25,35,38]

x = np.array([temp]).T   #T- Transporse method will do transverse # no cols will be the number of rows
# Linear regression method expects two dimensional column
print(x)
print(x.shape)

y = np.array(sale) #dependent or variable that needs to be predicted.
print(y)
print(y.shape)

```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
print(y_pred)
```

```python
# slope and intercept
print(model.coef_)
print(model.intercept_)
```

```python
# predicting new value
newoutput = model.predict([[80]])   #input is two dimensional array.
print(newoutput)
```

```python
mse = mean_squared_error(y,y_pred)
print(mse)
```

```python
plt.scatter(temp, sale, marker="*", color="r")
plt.plot(temp,y_pred, color='b')
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.title("Linear Regression")
plt.show()
```

```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(temp,sale)
temp = [20,25,30,35,40]
sale = [13,21,25,35,38]
print('slope ', slope)
print('intercept ', intercept)
print('r ', r)
print('p ', p)
print('std_err ', std_err)


def myfunc(x):
  return slope * x + intercept

y_pred = list(map(myfunc, temp))

plt.scatter(temp, sale)
plt.plot(temp,y_pred)
plt.show()

newsale = myfunc(45)
print(newsale)
"""
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.title("Linear Regression")
"""
```
