```python
from numpy.random import randn
import numpy as np

df = pd.DataFrame(randn(5,4),index= 'A B C D E'.split(),columns=['w','x','y','z'])
# randn function will create a random number. Split method will split the each index. index and columns names are case sensitive.

print(df)
```
